```python

# defining function to extract URL.

def webpage_dataframe(x):
    import requests
    import bs4
    import pandas as pd
    
    request=requests.get(x).text
    soup=bs4.BeautifulSoup(request,"html.parser")
    
    au=[]
    for i in soup.find_all("span",class_= "authorOrTitle"):
            au.append(i.text.strip())
            
    
    li=[]
    for i in soup.find_all("a",class_= "smallText"):
            li.append(i.text.strip())
    
    qu=[]
    for i in soup.find_all("div",class_= "quoteText"):
            c=list(i.children)
            qu.append(c[0].strip()) 
    
    data=pd.DataFrame({"author":au,"no_likes":li,"Qoutes":qu})
    return(data)
    
            
```

```python
# execution of above definitions

page_url=input()
webpage_dataframe(page_url)
```

```python
# new page

url= "http://www.goodreads.com/list/show/19.Best_for_Book_Clubs"
request= requests.get(url).text
soup=BeautifulSoup(request,"html.parser")

# extracting book title
def get_data(url):
    import bs4
    import pandas as pd
    import requests
    
    request= requests.get(url).text
    soup=bs4.BeautifulSoup(request,"html.parser")
    
    t=[]
    for i in soup.find_all("a",class_="bookTitle"):
        t.append(i.text.strip())
        

    r=[]
    for i in soup.find_all("span",class_="minirating"):
        c=list(i.children)
        r.append(c[1].split(" ")[1])
        
    author=[]
    for i in soup.find_all("a",class_="authorName"):
        c=list(i.children)
        author.append(i.text.strip())
        
    data1= pd.DataFrame({"Title":t,"Ratings":r,"Author":author})
    return data1
```
